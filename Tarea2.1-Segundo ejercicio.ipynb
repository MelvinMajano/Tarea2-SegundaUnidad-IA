```python
import pandas as pd

df = pd.read_csv('./housing.csv')
df.head()
```

```python
datos =df.dropna()
datos.isna().sum()
```

```python
dumies = pd.get_dummies(datos['ocean_proximity'], dtype= int)
datos = pd.concat([datos, dumies], axis=1)

datos.drop('ocean_proximity', axis=1, inplace=True)
datos
```

```python
datos['room_ratio'] = datos['total_rooms'] / datos['total_bedrooms']
datos['PupulaHouse'] = datos['population']/datos['households']
```

```python
EdadLimite =datos['housing_median_age']>50 #Verificamos que si hay datos que sean mayores a la censura que se nos dijo en el enunciado
EdadLimite
```

```python
PrecioLimite =datos['median_house_value']>500,000 #Verificamos que si hay datos que sean mayores a la censura que se nos dijo en el enunciado
EdadLimite
```

```python
IngresoLimite =datos['median_income']>15 #Verificamos que si hay datos que sean mayores a la censura que se nos dijo en el enunciado
IngresoLimite
```

```python
edadLimiteDumie = pd.get_dummies(datos['housing_median_age']>50, dtype= int, drop_first=True) #Como si hay datos que estan censurados decidmos hacer dumies de ellos
edadLimiteDumie.columns = ['edadLimiteDumie']
edadLimiteDumie
```

```python
PrecioLimiteDumie = pd.get_dummies(datos['median_house_value']>50000, dtype= int, drop_first=True) #Como si hay datos que estan censurados decidmos hacer dumies de ellos
PrecioLimiteDumie.columns = ['PrecioLimiteDumie']
PrecioLimiteDumie
```

```python
IngresoLimiteDumie = pd.get_dummies(datos['median_income']>15, dtype= int, drop_first=True) #Como si hay datos que estan censurados decidmos hacer dumies de ellos
IngresoLimiteDumie.columns = ['IngresoLimiteDumie']
IngresoLimiteDumie
```

```python
datos = pd.concat([datos,edadLimiteDumie,PrecioLimiteDumie,IngresoLimiteDumie], axis=1)  #Agregamos los dumies alas caracterizticas de entrada
```

```python
datos
```

```python
X=datos.drop(['median_house_value','ISLAND'],axis=1)
y=datos['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=9)

model = DecisionTreeRegressor(max_depth=20)

model.fit(X_train,y_train)

predicciones = model.predict(X_test)

comparativa = {"predicciones":predicciones, "valor real": y_test}
result = pd.DataFrame(comparativa)
result
```

```python
score = model.score(X_test, y_test)
score
```

```python
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt

plt.figure(figsize=(20,16))
plot_tree(
    model,
    feature_names=X_train.columns,
    class_names=['Acierto'],
    filled=True,
    label='none'
)
```
